# Training top models

```python
import pandas as pd
import warnings
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import tensorflow as tf

warnings.filterwarnings("ignore", category=UserWarning)

# Load dataset
dataset = pd.read_csv("C:/Users/samar/Documents/FINAL YEAR PROJECT FILES/Copy and Edit 2/datasets/Training.csv")

# Prepare X and y
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']

# Encode prognosis labels
le = LabelEncoder()
y = le.fit_transform(y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Standardize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# One-hot encode target labels for deep learning
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# Define individual models
svc = SVC(kernel='linear', probability=True)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
knn = KNeighborsClassifier(n_neighbors=5)
nb = MultinomialNB()

# Create Voting Classifier
voting_clf = VotingClassifier(estimators=[
    ('svc', svc),
    ('rf', rf),
    ('knn', knn),
    ('nb', nb)
], voting='soft')

# Train the voting classifier
voting_clf.fit(X_train, y_train)

# Evaluate the voting classifier
voting_predictions = voting_clf.predict(X_test)
voting_accuracy = accuracy_score(y_test, voting_predictions)
voting_cm = confusion_matrix(y_test, voting_predictions)

print(f"Voting Classifier Accuracy: {voting_accuracy*100}")
print("Voting Classifier Confusion Matrix:")
print(np.array2string(voting_cm, separator=', '))

# Adding the deep learning model
deep_model = Sequential()

# Input layer
deep_model.add(Dense(128, input_dim=X_train_scaled.shape[1], activation='relu'))

# Hidden layers
deep_model.add(Dense(64, activation='relu'))
deep_model.add(Dense(32, activation='relu'))

# Output layer
deep_model.add(Dense(len(le.classes_), activation='softmax'))

# Compile the model
deep_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the deep learning model
deep_model.fit(X_train_scaled, y_train_cat, epochs=50, batch_size=32, validation_data=(X_test_scaled, y_test_cat))

# Evaluate the deep learning model
deep_loss, deep_accuracy = deep_model.evaluate(X_test_scaled, y_test_cat)
print(f"Deep Learning Model Accuracy: {deep_accuracy*100}")

# Predict using the deep learning model
deep_predictions = deep_model.predict(X_test_scaled)
predicted_classes = np.argmax(deep_predictions, axis=1)

# Confusion matrix for deep learning model
deep_cm = confusion_matrix(y_test, predicted_classes)
print("Deep Learning Model Confusion Matrix:")
print(np.array2string(deep_cm, separator=', '))

# Convert predictions back to original labels for comparison
predicted_labels = le.inverse_transform(predicted_classes)

# Compare the results
print(f"\nComparison of Model Accuracies:")
print(f"Voting Classifier Accuracy: {voting_accuracy*100}")
print(f"Deep Learning Model Accuracy: {deep_accuracy*100}")

# Save the deep learning model
deep_model.save('combined_deep_learning_model.h5')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report
from nltk.translate.bleu_score import sentence_bleu

# Dictionary to store metrics for visualization
metrics = {'Model': [], 'Accuracy': [], 'F1-Score': [], 'Synthetic Score': [], 'BLEU Score': []}

# Function to calculate and collect metrics
def calculate_metrics(model_name, predictions):
    cm = confusion_matrix(y_test, predictions)
    accuracy = accuracy_score(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='weighted')
    synthetic_score = np.mean(predictions == y_test)  # Similar to accuracy but included for completeness
    bleu_score = sentence_bleu([list(y_test)], list(predictions))  # BLEU Score for comparison

    # Store metrics for visualization
    metrics['Model'].append(model_name)
    metrics['Accuracy'].append(accuracy)
    metrics['F1-Score'].append(f1)
    metrics['Synthetic Score'].append(synthetic_score)
    metrics['BLEU Score'].append(bleu_score)

    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'{model_name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Print classification report
    print(f"\n{model_name} - Classification Report:\n")
    print(classification_report(y_test, predictions))
    print(f"{model_name} Accuracy: {accuracy * 100:.2f}%")
    print(f"{model_name} F1 Score: {f1:.2f}")
    print(f"{model_name} Synthetic Score: {synthetic_score:.2f}")
    print(f"{model_name} BLEU Score: {bleu_score:.4f}")
    print("\n" + "=" * 40 + "\n")

# Calculate metrics for the Voting Classifier
calculate_metrics('Voting Classifier', voting_predictions)

# Calculate metrics for the Deep Learning Model
calculate_metrics('Deep Learning Model', predicted_classes)

# Create a DataFrame for visualization
df_metrics = pd.DataFrame(metrics)
df_metrics.set_index('Model', inplace=True)

# Plotting the metrics
df_metrics.plot(kind='bar', figsize=(12, 7), color=['#4CAF50', '#2196F3', '#FFC107', '#FF5733'])
plt.title('Model Performance Comparison')
plt.ylabel('Scores (0 to 1)')
plt.ylim(0, 1)  # Set y-axis limit to improve readability
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
