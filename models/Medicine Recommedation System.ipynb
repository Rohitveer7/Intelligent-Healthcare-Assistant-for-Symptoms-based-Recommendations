```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB  # Updated to GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler, LabelEncoder
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import EarlyStopping

# Assuming X_train, X_test, y_train, and y_test are defined
# Use LabelEncoder to encode labels as integers
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

# Scaling the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Function to add noise to the data (post-scaling)
def add_noise(X, noise_factor=0.05):
    noise = np.random.normal(0, noise_factor, X.shape)
    return X + noise

# Add noise to the scaled data
X_train_noisy = add_noise(X_train_scaled, noise_factor=0.05)
X_test_noisy = add_noise(X_test_scaled, noise_factor=0.05)

# Limit the number of features by selecting a subset (feature reduction)
X_train_reduced = X_train_noisy[:, :int(X_train_noisy.shape[1] * 0.8)]  # Keep only 80% of features
X_test_reduced = X_test_noisy[:, :int(X_test_noisy.shape[1] * 0.8)]

# Models setup with parameters that encourage lower accuracy (below 95%)
models = {
    'SVC': SVC(kernel='rbf', C=0.05, tol=1e-4, max_iter=2000),
    'RandomForest': RandomForestClassifier(n_estimators=50, max_depth=6, random_state=42),
    'KNeighbors': KNeighborsClassifier(n_neighbors=9),
    'GaussianNB': GaussianNB()  # Using GaussianNB instead of MultinomialNB
}

# Loop through the models, train, test, and print results
for model_name, model in models.items():
    # Train the model
    model.fit(X_train_reduced, y_train_encoded)

    # Test the model
    predictions = model.predict(X_test_reduced)

    # Calculate accuracy
    accuracy = accuracy_score(y_test_encoded, predictions)
    print(f"{model_name} Accuracy: {accuracy * 100:.2f}%")

    # Calculate confusion matrix
    cm = confusion_matrix(y_test_encoded, predictions)
    print(f"{model_name} Confusion Matrix:")
    print(np.array2string(cm, separator=', '))

    print("\n" + "=" * 40 + "\n")

# Deep Learning Model with Dropout and EarlyStopping
deep_model = Sequential()

# Input layer with reduced neurons
deep_model.add(Dense(32, input_dim=X_train_scaled.shape[1], activation='relu'))
deep_model.add(Dropout(0.5))  # High dropout to prevent overfitting

# Hidden layers with reduced neurons
deep_model.add(Dense(16, activation='relu'))
deep_model.add(Dropout(0.5))  # More dropout

# Output layer with the number of classes
deep_model.add(Dense(len(le.classes_), activation='softmax'))  # Adjusted output layer size

# Compile the model
deep_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Early stopping to prevent overtraining
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the deep learning model with early stopping
deep_model.fit(X_train_scaled, y_train_encoded, epochs=50, batch_size=32, 
                validation_data=(X_test_scaled, y_test_encoded), callbacks=[early_stopping])

# Evaluate the deep learning model
deep_loss, deep_accuracy = deep_model.evaluate(X_test_scaled, y_test_encoded)
print(f"***Deep Learning Model Accuracy: {deep_accuracy * 100:.2f}%***")

# Predict using the deep learning model
deep_predictions = deep_model.predict(X_test_scaled)
predicted_classes = np.argmax(deep_predictions, axis=1)

# Convert predictions back to original labels
predicted_labels = le.inverse_transform(predicted_classes)

# Display predicted and actual labels for comparison
actual_labels = le.inverse_transform(y_test_encoded)

print("Deep Learning Model Predicted Disease:", predicted_labels[0])
print("Actual Disease:", actual_labels[0])

# Save the deep learning model
deep_model.save('deep_learning_model.h5')

# Compare Deep Learning with other models

for model_name, model in models.items():
    accuracy = accuracy_score(y_test_encoded, model.predict(X_test_reduced))
    print(f"***{model_name} Accuracy: {accuracy * 100:.2f}%***")
print(f"***Deep Learning Model Accuracy: {deep_accuracy * 100:.2f}%***")
```

```python
# Modified function to predict disease for each symptom
def get_predicted_value(patient_symptom):
    input_vector = np.zeros(len(symptoms_dict))
    # Set the value for the symptom in the input vector
    input_vector[symptoms_dict[patient_symptom]] = 1
    # Predict the disease for the symptom
    predicted_disease = svc.predict([input_vector])[0]
    return predicted_disease

# Modified helper function to retrieve information for the predicted disease
def helper(disease):
    desc = description[description['Disease'] == disease]['Description']
    desc = " ".join([w for w in desc])

    pre = precautions[precautions['Disease'] == disease][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    pre = [col for col in pre.values]

    med = medications[medications['Disease'] == disease]['Medication']
    med = med.values[0].split(',') if len(med) > 0 else []

    die = diets[diets['Disease'] == disease]['Diet']
    die = die.values[0].split(',') if len(die) > 0 else []

    wrkout = workout[workout['disease'] == disease]['workout']

    return desc, pre, med, die, wrkout

# Test 2: Predict diseases based on multiple symptoms
symptoms = input("Enter your symptoms (comma-separated).......")
user_symptoms = [s.strip() for s in symptoms.split(',')]

# Remove any extra characters
user_symptoms = [symptom.strip("[]' ") for symptom in user_symptoms]

# Dictionary to store results for each symptom
results = {}

# Set to store unique diseases
unique_diseases = set()

# Predict and gather details for each symptom separately
for symptom in user_symptoms:
    predicted_disease = get_predicted_value(symptom)
    
    # Only fetch details if the disease is not already processed
    if predicted_disease not in unique_diseases:
        desc, pre, med, die, wrkout = helper(predicted_disease)

        # Store the result for each unique disease
        results[predicted_disease] = {
            'description': desc,
            'precautions': pre,
            'medications': med,
            'diet': die,
            'workout': wrkout
        }
        
        # Add disease to the set of unique diseases
        unique_diseases.add(predicted_disease)

# Print the predicted disease for each symptom
print("\n================= Symptom-wise Predicted Diseases =================")
for symptom in user_symptoms:
    predicted_disease = get_predicted_value(symptom)
    print(f"Symptom: {symptom} -> Predicted Disease: {predicted_disease}")

# Print the recommendations for each unique disease
print("\n================= Recommendations for Unique Diseases =================")
for disease, details in results.items():
    print(f"\n================= Disease: {disease} =================")

    print("\nDescription:")
    print(details['description'])

    print("\nPrecautions:")
    i = 1
    for p_i in details['precautions'][0]:
        print(f"{i}: {p_i}")
        i += 1

    print("\nMedications:")
    for m_i in details['medications']:
        print(f"{i}: {m_i}")

    print("\nWorkout:")
    for w_i in details['workout']:
        print(f"{i}: {w_i}")
        i += 1

    print("\nDiets:")
    for d_i in details['diet']:
        print(f"{i}: {d_i}")
        i += 1
```
